```python
active = group58.groupby(['date'])['clickrate_active'].sum().to_frame('sum_click_active').reset_index()
```

```python
active.drop('date',axis=1,inplace=True)
```

```python
active
```

```python
plt.figure(figsize=(12, 5))
active['sum_click_active'].plot(label='active',color='blue')
```

##Apply ARIMA model to active learning style

```python
adfuller(active.sum_click_active.diff().dropna())
```

```python
arima_model = sm.tsa.arima.ARIMA(active.sum_click_active[:],order=(5,1,5))
model = arima_model.fit()
print(model.summary())
```

```python
from statsmodels.graphics.tsaplots import plot_predict
plot_predict(model, 0,100)
```

```python
pred = model.forecast(3)
```

```python
pred = pd.DataFrame(pred)
```
