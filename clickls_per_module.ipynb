```python
import pandas as pd
import numpy as np
import random
from numpy.random import randint
from numpy.random import rand

import warnings
warnings.filterwarnings('ignore')
from collections import ChainMap
```

```python
clickrate = pd.read_csv("/content/drive/MyDrive/PCD/Colab Notebooks/clickrate_cat_ls.csv")
```

```python
clickrate.drop(["Unnamed: 0"], axis = 1,inplace = True)
```

```python
clickrate.head()
```

```python
kmeans_table1 = clickrate.groupby(['id_student','code_module','code_presentation']).apply(lambda x: x[x["activity_cat_ls"] == "reflective" ]['sum_click'].sum()).reset_index()
kmeans_table2 = clickrate.groupby(['id_student','code_module','code_presentation']).apply(lambda x: x[x["activity_cat_ls"] == "global" ]['sum_click'].sum()).reset_index()
kmeans_table3 = clickrate.groupby(['id_student','code_module','code_presentation']).apply(lambda x: x[x["activity_cat_ls"] == "sensing" ]['sum_click'].sum()).reset_index()
kmeans_table4 = clickrate.groupby(['id_student','code_module','code_presentation']).apply(lambda x: x[x["activity_cat_ls"] == "verbal" ]['sum_click'].sum()).reset_index()
kmeans_table5 = clickrate.groupby(['id_student','code_module','code_presentation']).apply(lambda x: x[x["activity_cat_ls"] == "visual" ]['sum_click'].sum()).reset_index()
kmeans_table6 = clickrate.groupby(['id_student','code_module','code_presentation']).apply(lambda x: x[x["activity_cat_ls"] == "sequential" ]['sum_click'].sum()).reset_index()
```

```python
kmeans_table0 = clickrate.groupby(['id_student','code_module','code_presentation']).apply(lambda x: x[x["activity_cat_ls"] == "active" ]['sum_click'].sum()).reset_index()
```

```python
kmeans_table1.rename(columns = {0:'clickrate_reflective'}, inplace = True)
kmeans_table2.rename(columns = {0:'clickrate_global'}, inplace = True)
kmeans_table3.rename(columns = {0:'clickrate_sensing'}, inplace = True)
kmeans_table4.rename(columns = {0:'clickrate_verbal'}, inplace = True)
kmeans_table5.rename(columns = {0:'clickrate_visual'}, inplace = True)
kmeans_table0.rename(columns = {0:'clickrate_active'}, inplace = True)
kmeans_table6.rename(columns = {0:'clickrate_sequential'}, inplace = True)
```

```python
test1 = clickrate.groupby(['id_student','code_module'])['sum_click'].sum().reset_index()
test1.head(10)
```

```python
kmeans_table0['clickrate_active'] = kmeans_table0['clickrate_active']/test1['sum_click']
kmeans_table1['clickrate_reflective'] = kmeans_table1['clickrate_reflective']/test1['sum_click']
kmeans_table2['clickrate_global'] = kmeans_table2['clickrate_global']/test1['sum_click']
kmeans_table3['clickrate_sensing'] = kmeans_table3['clickrate_sensing']/test1['sum_click']
kmeans_table4['clickrate_verbal'] = kmeans_table4['clickrate_verbal']/test1['sum_click']
kmeans_table6['clickrate_sequential'] = kmeans_table6['clickrate_sequential']/test1['sum_click']
kmeans_table5['clickrate_visual'] = kmeans_table5['clickrate_visual']/test1['sum_click']
```

```python
kmeans_table0.head()
```

```python
test=pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(kmeans_table0,kmeans_table1,on=['id_student','code_module','code_presentation']),kmeans_table2,on=['id_student','code_module','code_presentation']),kmeans_table3,on=['id_student','code_module','code_presentation']),kmeans_table4,on=['id_student','code_module','code_presentation']),kmeans_table4,on=['id_student','code_module','code_presentation']),kmeans_table5,on=['id_student','code_module','code_presentation']),kmeans_table6,on=['id_student','code_module','code_presentation'])
```
